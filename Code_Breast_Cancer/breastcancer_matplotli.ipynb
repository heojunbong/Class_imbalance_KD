```python
# 필요한 라이브러리 긁어오기

import pandas as pd 
import matplotlib.pyplot as plt 
```

```python

Experiments_case=['Baseline','Case1','Case2','Case3']
import numpy as np 
index=np.arange(4)
Accuracy=[0.7193, 0.7975,0.7851,0.8025]
Precision=[0.2809,0.3413,0.3282,0.3475]
Recall=[0.5366,0.3496,0.3415,0.3333]
F1_score=[0.5941,0.6128,0.5995,0.6121]
Auc_score=[0.6988,0.6902,0.6861,0.6910]
```

```python
bar_width=0.15
fig,ax=plt.subplots(figsize=(18,6))
b1=plt.bar(index,Accuracy,bar_width,alpha=0.4,color='red',label='accuracy')
b2=plt.bar(index+bar_width,Precision,bar_width,alpha=0.4,color='blue',label='precision')
b3=plt.bar(index+2*bar_width,Recall,bar_width,alpha=0.4,color='green',label='recall')
b4=plt.bar(index+3*bar_width,F1_score,bar_width,alpha=0.4,color='yellow',label='f1_score')
b5=plt.bar(index+4*bar_width,Auc_score,bar_width,alpha=0.4,color='purple',label='AUC_Score')

plt.xticks(index,Experiments_case)
plt.legend(fontsize=5)
plt.xlabel('< Experiment_Case >')
plt.ylabel('Metric Score')
plt.show()
```
